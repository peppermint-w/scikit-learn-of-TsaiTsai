### 一、处理分类型特征：编码与哑变量

将文字型数据转换为数值型。

1.**sklearn.preprocessing.LabelEncoder**：能够将分类**标签**转换为分类数值

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y = data.iloc[:,-1] #要输入的是标签，不是特征矩阵，所以允许一维
```

```python
le = LabelEncoder() #实例化
le = le.fit(y) #导入数据
label = le.transform(y) #transform接口调取结果
label
```

```python
le.classes_ #以下标签分别替换为0、1、2
```

```python
le.fit_transform(y) #后两步一步到位
```

```python
le.inverse_transform(label) #使用inverse_transform可以逆转
```

```python
data.iloc[:,-1] = label #让标签等于运行出来的结果
```

```python
data.head()
```

如果不需要教学展示会这么写：<br>
data.iloc[:,-1] = LabelEncoder().fit_transform(data.iloc[:-1])

2.**sklearn.preprocessing.OrdinalEncoder**：能够将分类**特征**（有序变量）转换为分类数值

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data_ = data.copy()
data_.head()
```

```python
X = data_.iloc[:,1:-1]
```

```python
fe = OrdinalEncoder() #实例化
fe = fe.fit(X) #导入数据
features = fe.transform(X) #transform接口调取结果
features
```

```python
fe.categories_ #类似classes_
```

```python
data_.iloc[:,1:-1] = features
data_.head()
```

如果不需要教学展示会这么写：<br>
data.iloc[:,1:-1] = OrdinalEncoder().fit_transform(data.iloc[:,1:-1])

注意，**性别**和**舱门**是**名义变量**（各取值之间相互独立，彼此之间完全没有联系），这种处理方式在算法看来，是连续且可以计算的，事实上并非如此，因此要使用`OneHotEncoder`哑变量的方式来处理。对于**学历**这种**有序变量**可以使用`OrdinalEncoder`来处理。

`OrdinalEncoder`：["S","Q","C"]→[0,1,2]<br>
`OneHotEncoder`：["S","Q","C"]→[[1,0,0],[0,1,0],[0,0,1]]

3.**sklearn.preprocessing.OneHotEncoder**：能够将分类**特征**（名义变量）转换为分类数值

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
X = data.iloc[:,1:-1]
X.head()
```

```python
enc = OneHotEncoder(categories='auto').fit(X)
result = enc.transform(X).toarray()
result
```

result = OneHotEncoder(categories='auto').fit_transform(X).toarray()

```python
enc.get_feature_names() #每个哑变量的类别
```

```python
#依然可以还原
pd.DataFrame(enc.inverse_transform(result))
```

```python
#将两个表左右相连
newdata = pd.concat([data,pd.DataFrame(result)],axis=1)
newdata
```

```python
#删除Sex和Embarked列，并覆盖原数据
newdata.drop(["Sex","Embarked"],axis=1,inplace=True)
```

```python
newdata.head()
```

```python
#改列索引
newdata.columns = ['Age','Survived','Female','Male','Embarked_C','Embarked_Q','Embarked_S']
newdata.head()
```

**5.总结。**

编码与哑变量|功能|重要参数|重要属性|重要接口
-|-|-|-|-
.LabelEncoder|分类标签编码|N/A|.classes_：查看标签中有多少类别|fit,<br>transform,<br>fit_transform,<br>inverse_transform
.OrdinalEncoder|分类特征编码|N/A|.categories_：查看特征中有多少类别|fit,<br>transform,<br>fit_transform,<br>inverse_transform
.OneHotEncoder|独热编码，为名义变量创建哑变量|categories='auto'算法自己判断每个特征有哪些类别|.categories_：查看特征中有多少类别|fit,<br>transform,<br>fit_transform,<br>inverse_transform,<br>get_feature_names

### 二、处理连续型数据：二值化与分段

1.**sklearn.preprocessing.Binarizer**：二值化

根据阈值将数据二值化，用于处理连续型变量。大于阈值的值映射为1，小于或等于阈值的值映射为0。

（1）年龄二值化：阈值30

```python
from sklearn.preprocessing import Binarizer
```

```python
data_2 = data.copy()
X = data_2.iloc[:,0].values.reshape(-1,1) #类为特征专用，所以不能使用一维数组
```

```python
transformer = Binarizer(threshold=30).fit_transform(X) #三步一步到位，大于等于30岁用1替换
transformer
```

```python
data_2.iloc[:,0] = transformer
data_2.head()
```

（2）标签二值化：阈值1，即unknown和死亡都用死亡替换

2.**sklearn.preprocessing.KBinsDiscretizer**分箱：将连续型变量划分为分类变量，将连续型变量排序后按顺序分箱后编码。

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
X = data.iloc[:,0].values.reshape(-1,1)
```

```python
est = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform') #实例化，每一列是一个特征
est.fit_transform(X) #后两步合一步
```

```python
#查看转换后分的箱：变成了一列中的三箱
set(est.fit_transform(X).ravel()) #降成一维放在集合中
```

```python
est = KBinsDiscretizer(n_bins=3, encode='onehot', strategy='uniform') #实例化，每一列是一个特征中的一个类别
```

```python
#查看转换后分的箱：变成了哑变量
est.fit_transform(X).toarray()
```
