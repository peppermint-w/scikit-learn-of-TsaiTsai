```python
import pandas as pd
```

```python
data = pd.read_csv('02Narrativedata.csv',index_col=0) #原数据第1列有索引，index_col=0把第1列作为索引，而不是特征
data.head()
```

**1.使用sklearn.impute.SimpleImputer填充年龄（数值型数据）**

```python
data.loc[:,"Age"] #loc使用索引名字切片，返回Series
```

```python
data.loc[:,"Age"].values
```

```python
data.loc[:,"Age"].values.shape #是一维，需要升维
```

```python
data.loc[:,"Age"].values.reshape(-1,1) #升维
```

```python
data.loc[:,"Age"].values.reshape(-1,1).shape #已升维
```

```python
Age = data.loc[:,"Age"].values.reshape(-1,1)
```

```python
from sklearn.impute import SimpleImputer
```

```python
#实例化
imp_mean = SimpleImputer() #用均值填补
imp_median = SimpleImputer(strategy="median") #用中位数填补
imp_0 = SimpleImputer(strategy="constant",fill_value=0) #用0填补
```

```python
#用fit_transform调取结果
imp_mean = imp_mean.fit_transform(Age)
imp_median = imp_median.fit_transform(Age)
imp_0 = imp_0.fit_transform(Age)
```

```python
imp_mean #均值是小数，且与中位数接近
```

```python
imp_median #选择中位数
```

```python
imp_0
```

```python
data.loc[:,"Age"] = imp_median
data.info()
```

**2.使用sklearn.impute.SimpleImputer填补Embarked（分类型数据）**

```python
Embarked = data.loc[:,"Embarked"].values.reshape(-1,1)
```

```python
imp_mode = SimpleImputer(strategy = "most_frequent") #用众数填补，现实案例可删除，因为只有2条
data.loc[:,"Embarked"] = imp_mode.fit_transform(Embarked)
data.info()
```

**3.使用pandas填补缺失值**

```python
data_ = pd.read_csv('02Narrativedata.csv',index_col=0) #原数据第1列有索引，index_col=0把第1列作为索引，而不是特征
data_.info()
```

```python
data_.loc[:,"Age"] = data_.loc[:,"Age"].fillna(data_.loc[:,"Age"].median())
data_.dropna(axis=0, inplace=True) #删除所有缺失值的行，axis=1删除所有缺失值的列
```

```python
data_.info()
```
